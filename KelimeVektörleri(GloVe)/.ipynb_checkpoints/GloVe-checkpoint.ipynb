<center> Kelime Sayma(SVD) </center>





- svd eski bir method 
- matrix büyüdükce bilgisayar yükü cok fazla büyüyor
- n x m matris icin : O(mn2)
- milyonlarca kelime icin kötü
- yeni kelime veya dökuman eklemek zor
- derin ögrenmeden faydalanmiyor
- matrix cok büyük olmadikca hizli

Global Vectors (GloVe) 
- stanford unideki arastirmacilar gelistiriyor
- kelime sayma ile w2v avantajlarini kullanir
- hizli egitim
- cok büyük corpuslardada kullanilir
- kücük corpus kücük vektörlerde de iyi performans
- w2v e göre daha hizli
- w2v(skip gram): corpus u kelime kelime gezerek iki kelimenin ne kadar sik yan yana oldugunu hesapliyor. kelime kelime gezerken optimizasyon da yapiliyor.
- GloVe ise iki kelimenin ne kadar yan yana oldugunun bir istatistigini cikartiyr. Istatistik elde edildikten sonra optimizasyon yapiliyor
