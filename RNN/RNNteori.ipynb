- cizgiler ayriliyorsa: vektr kopyalaniyor
- cizgiler birlesiyorsa: iki vektr birlesiyor
- kirmizi dairelerde: carpma ve ya toplama
- dikdrtgen kutularda: 3 nde sigmoid 1 inde tanh aktivasyon func kullaniliyro
Bilgi akisinin saglandigi bazi gecitler var:
1.) cell state: hcre durumunda tasinacak bilgi. heradimda gncellenir. 
2.) forget gate: lstm kendisine gelen bilgilerden hangisini unutmasi gerektigine karar veriyor. bunu sigmoid(0-1) ile yapiyor.
3.) Input gate: cell state de hangi degerlerin saklanmasi gerektigini belirler. Hangi degerlerin gncellenmesi gerektigini belirler. tanh de ise cell state eklenmeye aday olan yeni degerlerden bir vektr olusturuluyor. Daha sonra bu ikisi birlestirilip cell state e eklenir.
4.) cell state gncelleme: Bu gncelleme ilemi, veriye bal olarak hcre durumuna ne kadar bilgi eklenip karlacan belirler.
5.) k Kaps (Output Gate): Hcre durumu zerinden kan bilgiler, son tahminlerin oluturulmasnda kullanlr. k kaps, hangi bilgilerin dar kacan kontrol eder.
