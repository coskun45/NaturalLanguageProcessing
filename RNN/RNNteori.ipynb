# Rnn in Yapisi

- bir input ve önceki hidden state i alip yeni bir output üretiyor

# Long Short TErm Memory(LSTM)

- LSTM'nin temel amacı, uzun süreli bağımlılıkları modellemek ve sıralı verilerdeki önemli bilgileri saklamaktır.
- vanishing Gradient proplemi ni cözer
- basit RNN e göre daha effizient.



- cizgilerden vektörlerden akiyor
- cizgiler ayriliyorsa: vektör kopyalaniyor
- cizgiler birlesiyorsa: iki vektör birlesiyor
- kirmizi dairelerde: carpma ve ya toplama
- dikdörtgen kutularda: 3 ünde sigmoid 1 inde tanh aktivasyon func kullaniliyro
Bilgi akisinin saglandigi bazi gecitler var:
1.) cell state: hücre durumunda tasinacak bilgi. heradimda güncellenir. 
2.) forget gate: lstm kendisine gelen bilgilerden hangisini unutmasi gerektigine karar veriyor. bunu sigmoid(0-1) ile yapiyor.
3.) Input gate: cell state de hangi degerlerin saklanmasi gerektigini belirler. Hangi degerlerin güncellenmesi gerektigini belirler. tanh de ise cell state eklenmeye aday olan yeni degerlerden bir vektör olusturuluyor. Daha sonra bu ikisi birlestirilip cell state e eklenir.
4.) cell state güncelleme: Bu güncelleme işlemi, veriye bağlı olarak hücre durumuna ne kadar bilgi eklenip çıkarılacağını belirler.
5.) Çıkış Kapısı (Output Gate): Hücre durumu üzerinden çıkan bilgiler, son tahminlerin oluşturulmasında kullanılır. Çıkış kapısı, hangi bilgilerin dışarı çıkacağını kontrol eder.


# Grated Recurrent Unit(GRU)
- bir lstm cesidi
- gru da cell state yok. yerine hidden state var.
- lstm de cell ve hidden state var. gru da sadece hidden state var.
- meory ve hidden state icin 2 ayri deger olusturmak yerine tek state kullanilr
- forget ve input gate i birlestirip tek bir update gate kullaniyorz
