- bu sorunu cözmek icin kelimeleri temsil eden bir vektör olusturacagiz.
- her kelime icin bir vektör olusturulacak
- vektörler kelimeler arasindaki anlamasal baglantiyida saklayacak
- birbirine benzer kelimelerin vektörleride birbirine yakin olack
- kelime vektörü olusturmak icin en modern yöntem word2vec,GloVe
- bu sayede kelimeler arasinda matematiksel islemler yapabilirz.

<center> <h2><b>  Word2vec </b> </h2> </center>

- kelimenin etrafindaki diger kelimeler o kelimeyi temsil eder
- wikipedia gibi bir corpus umuz varsa dogru vektörler olusturabiliriz
- gergedan kelimesini icin dogru vektörü olusturmak istiyorsak gergedan kullanildigi bütün cümleleri bilmak gerekir.
- insanda benzer sekilde kelimeleri anlar. kelimenin etrafindaki kelimelerden kelimenin anlamini cikartma
- 2013 yilinda google tarafindan bulunmus gözetimsiz ögrenme

- ortadaki kelimeden etrafindakileri tahmin etmeye calisan bir model olusturuyoruz 
- elimizdeki yazida bulunan bütün kelimelerin bir vektörü olacak
- ortadaki kelimenin etrafindaki kelimeler icin bazi olasilik degerleri alacagiz. Bu olasilik degerleri icin bir loss fonksiyonu bulacagiz ve amac bu loss func minimize etmek.
- kelime vektörlerini train esnasinda degistirerek loss u minimize etmeye calisiyoruz
- w2v icinde 2 algoritma var

<center> <h2>negativ sampling </h2> </center>

- Word2vec modelleri, bir giriş kelimesine göre çevresindeki veya bağlamdaki kelimeleri tahmin etmeye çalışır. Bu, çok sınıflı bir sınıflandırma problemidir ve tüm kelime dağılımını tahmin etmek için softmax fonksiyonu kullanılır. Ancak softmax fonksiyonu, çok büyük bir çıktı katmanına uygulandığında çok maliyetlidir.

- Negative sampling, bu sorunu çözmek için softmax fonksiyonunun yerine ikili lojistik regresyon (ikili sınıflandırma) kullanmayı amaçlayan yeni bir hedef fonksiyon tanımlar. Her eğitim örneği için, iyileştirilmiş (optimize edilmiş) sınıflandırıcıya gerçek bir çift (bir merkez kelime ve bağlamında görünen başka bir kelime) ve k tane rastgele bozulmuş çift (merkez kelime ve sözlükten rastgele seçilen bir kelime) verilir. Sınıflandırıcı gerçek çiftleri bozulmuş olanlardan ayırt etmeyi öğrenerek, sonunda kelime vektörlerini de öğrenir12
